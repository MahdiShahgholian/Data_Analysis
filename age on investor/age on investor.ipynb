```python
dm = df.copy()
nonumerical = ["Sex", "Occupation", "Education Level", "Proportion Invest", "Invest Stocks", "Factors Investment", "Investment Objective", "Duration", "Expected Return", "Reason Equity", "Reason MF", "Reason Gsec", "Reason FD", "Source"]
encoder = LabelEncoder()
for i in nonumerical:
    dm[i] = encoder.fit_transform(dm[i])
dm.corr()
```

```python
plt.figure(figsize=(30,30))
sns.heatmap(dm.corr(), annot=True, cmap='coolwarm', linewidth=0.6, )
plt.title('Correlation Matrix')
```
